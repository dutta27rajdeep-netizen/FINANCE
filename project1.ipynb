```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import os
```

## **DEFINE TICKER AND DATE RANGE**

Define the list of tickers

```python
tickers = ["RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "INFY.NS", "ICICIBANK.NS",
           "WIPRO.NS", "BHARTIARTL.NS", "SBIN.NS", "HINDUNILVR.NS", "KOTAKBANK.NS"]

```

```python
end_date = datetime.now()
print(end_date)

```

```python
start_date = end_date - timedelta(days=3*365)
print(start_date)
```

```python
close_df = pd.DataFrame()
```

```python
for ticker in tickers:
    data = yf.download(ticker, start="2023-01-01", end=end_date)
    close_df[ticker] = data["Close"]

print(close_df)
```

```python
output_folder = r"C:\ONEDRIVE\Desktop\CAREER\PYTHON\Finance\FINANCE"
```

```python
output_file = os.path.join(output_folder, "close_prices.xlsx")
close_df.to_excel(output_file)
```
